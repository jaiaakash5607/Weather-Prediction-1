```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv("/content/play_tennis_50.csv")
df.head()
```

```python
numerics=LabelEncoder()
```

```python
input_d=df.drop('Play',axis='columns')
target=df['Play']
target.head()
```

```python
input_d['Outlook_n']=numerics.fit_transform(input_d['Outlook'])
input_d['Temp_n']=numerics.fit_transform(input_d['Temp'])
input_d['Humidity_n']=numerics.fit_transform(input_d['Humidity'])
input_d['Windy_n']=numerics.fit_transform(input_d['Windy'])

input_d.head(10)
```

```python
input_d=input_d.drop(['Outlook','Temp','Humidity','Windy'],axis="columns")
input_d.head()
```

```python
Classifier=GaussianNB()
Classifier.fit(input_d,target)
```

```python
Classifier.score(input_d,target)
```

```python
Out=int(input("enter the Outlook | 0.Overcast | 1.Rainy | 2.Sunny :"))
Tem=int(input("enter the Temp | 0.Cool | 1.HOT | 2.Mild :"))
Hum=int(input("enter Humidity | 0.High | 1.Normal :"))
Win=int(input("enter Windy | 0.F | 1.T :"))

print(Classifier.predict([[Out,Tem,Hum,Win]]))
```
